```python
### REMOVE OUTLIERS FOR COLUMN ~ 'Added Sugars' ~

q1_added = train_cleaned["added_sugars"].quantile(0.25)
q3_added = train_cleaned["added_sugars"].quantile(0.75)
#Interquartile Range is Q3 - Q1
IQR_added = q3_added - q1_added
#minimum outliers and maximum outliers
min_cond_added = q1_added - (1.5 * IQR_added) 
max_cond_added = q3_added + (1.5 * IQR_added)
#Condition masks for those two outliers
min_outliers_added = train_cleaned["added_sugars"] < min_cond_added
max_outliers_added = train_cleaned["added_sugars"] > max_cond_added
#select datas inverse the condtions => take all the datas but those satisfied the condition masks which are outliers
removed_added = train_cleaned[~(min_outliers_added | max_outliers_added)]
### REMOVE OUTLIERS FOR COLUMN ~ 'Protein' ~

q1_protein = train_cleaned["protein"].quantile(0.25)
q3_protein = train_cleaned["protein"].quantile(0.75)
#Interquartile Range is Q3 - Q1
IQR_protein = q3_protein - q1_protein
#minimum outliers and maximum outliers
min_cond_protein = q1_protein - (1.5 * IQR_protein) 
max_cond_protein = q3_protein + (1.5 * IQR_protein)
#Condition masks for those two outliers
min_outliers_protein = train_cleaned["protein"] < min_cond_protein
max_outliers_protein = train_cleaned["protein"] > max_cond_protein
#select datas inverse the condtions => take all the datas but those satisfied the condition masks which are outliers
removed_protein = train_cleaned[~(min_outliers_protein | max_outliers_protein)]
removed_outliers = removed_protein + removed_added
removed_outliers["product_type"] = removed_outliers["product_type"].str.replace("FOODFOOD", "FOOD")
removed_outliers["product_type"] = removed_outliers["product_type"].str.replace("BEVERAGEBEVERAGE", "BEVERAGE")
```

```python

# make figure and assign axis objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 15))

fig.subplots_adjust(wspace=0)

# pie chart parameters
overall_ratios = train_cleaned["country"].value_counts()/len(train_cleaned["country"])
labels = train_cleaned["country"].value_counts().index

# rotate so that first wedge is split by the x-axis
angle = -180 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, autopct='%1.1f%%', startangle=angle,
                     labels=labels, explode = explode)

# bar chart parameters
product_ratios = train_cleaned["product_type"].value_counts()/len(train_cleaned["product_type"])
product_labels = train_cleaned["product_type"].value_counts().index
bottom = 1
width = .2

# Adding from the top matches the legend.
for j, (height, label) in enumerate(reversed([*zip(product_ratios, product_labels)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='royalblue', label=label,
                 alpha=0.1 + 0.25 * j)
    ax2.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')

ax2.set_title('Percentage of product type in the most occupied country')
ax2.legend()
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)

# use ConnectionPatch to draw lines between the two plots
theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(product_ratios)

# draw top connecting line
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(-width / 2, bar_height), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(4)
ax2.add_artist(con)

# draw bottom connecting line
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(-width / 2, 0), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(4)

plt.show()
```

```python
#HANDLING WITH EMPTY VALUES IN TRAINING DATASET

train_cleaned = train_cleaned.fillna(-999)
test_cleaned = test_cleaned.fillna(-999)

#CREATING A CONSTANT VALUE COLUMN IN ORDER TO MAKE THE MODEL NON-NEGATIVE
train_cleaned["constant"] = 1
test_cleaned["constant"] = 1

```

```python
#GET DUMMIES VARIABLES ON CATEGORY TO GET SUGARS BASED ON IT
train_cleaned = pd.get_dummies(train_cleaned, columns= ["category_name", "product_type"], drop_first = True)
test_cleaned = pd.get_dummies(test_cleaned, columns= ["category_name", "product_type"], drop_first = True)
```

```python
#SELECT FEATURES TO TRAIN THE MODEL

feature_selected = ["total_calories", "total_fat", "sugars", 
                     "protein",'%_of_energy_from_added_sugars', "cholesterol",
                    'sodium', 'sodium_per_kcal', 'constant',
                    'dietary_fiber','category_name_CONFECTIONARY', 'category_name_COOKIES & CRACKERS',
                   'category_name_DAIRY', 'category_name_DAIRY FOODS', 'category_name_FOODS',
                   'category_name_GRAINS', 'category_name_NUTRITION',
                   'category_name_SAVOURY SNACKS'
                   ]
```

```python
#FIT THE TRAINING DATA TO TRAIN THE MODEL 80% DATASET FOR TRAINING DATA, 20% FOR TESTING DATASET

x_train = train_cleaned[feature_selected]
y_train = train_cleaned["added_sugars"]
x_test = test_cleaned[feature_selected]
y_test = test_cleaned["added_sugars"]
```

```python
#PICK RandomForestRegressor to be a predictive model

model = RandomForestRegressor(n_estimators = 1000)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test).round(2)
```

```python
#MAKE A DATAFRAME OF PREDICTIONS AND ACTUAL TO SEE THE PRECISION AND ACCURACY OF THE MODEL

pred_df = pd.DataFrame(y_pred, columns = ["predictions"])
pred_df["actual"] = y_test
pred_df["abs"] = abs(pred_df["predictions"] - pred_df["actual"])
pred_df.head()
```

```python
#R-Squared Score OF THE PREDICTIONS AND ACTUAL THAT ACQUIRED FROM THE MODEL

r_squared = metrics.r2_score(y_test, y_pred)
print('R-Squared Score:', r_squared)

# USE SKLEARN'S MEAN_ABSOLUTE_ERROR FUNCTION TO SCORE OUR PREDICTIONS VS OUR ACTUAL
mae = metrics.mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```
